### Tutorial outcomes

### Interface Overview
In the previous tutorials, we found that the AXILite is the simplest interface, but performance is lower. Streams improve performance, but require a DMA to transfer data. AXI Master can be used to access data in DDR DRAM directly without needing a DMA. This can simplify the hardware and software design, and can increase performance as some of the control is offloaded from the PS to the IP.

### Changes in Code

In order to copy data to and from memory, we can use the function `memcpy` in the HLS Code. This function is only sythesisable in this situation because the interface type for the arrays we are copying is <i>m_axi</i>; in most other cases, use of this function will generate an error. In order to pipeline the design fully, and avoid using a significant amount of memory in the PL, data will actually be copied 128 bits at a time. This is only possible because the input and output have been changed to arrays of type `ap_uint<128>`. If we had kept arrays, the data would be copied one byte at a time, which would increase the throughput to 16. Despite the different type, the _size_ argument of `memcpy` is still in bytes. 

Most functions called in the top level function are contained in a loop, which will act on each incoming block of data. The functions `read_key` and `KeyExpansion` are kept outside the loop as they only need to be executed once, as this mode of AES only uses one key.

### Directives 
By pipelining the main loop, we allow all the functions called within the loop to be pipelined, and any loops within them unrolled. This allows this loop to have a throughput of 1. We also need a _Dataflow_ directive here to allow `memcpy`, `read_key` and `KeyExpansion` to be scheduled more optimally. These directives are already included in the source code, but the interfaces still need to be set:

* `key` and `size` are both simple scalar inputs, so we can use an AXILite interface. Double click on each of these and select the mode `s_axilite`. These directives will appear in the directives view, or you can decide to make these appear in the source file as
```C 
#pragma HLS INTERFACE s_axilite port=key
#pragma HLS INTERFACE s_axilite port=size
``` 
* `input` and `output` will be accessed via the master interfaces here, using the mode <i>m_axi</i>. The _depth_ must be set, but is only used for Co-Simulation. You can set this to the number of bytes used for the input in the test bench, or slightly greater. The _offset_ option determines how we will determine the physical address of the arrays in DDR memory. Set this to _slave_ to create a register within our existing AXILite bundle dedicated to holding the address offset of the arrays. In the source file, these will appear as 
```C
#pragma HLS INTERFACE m_axi depth=160 port=input offset=slave
#pragma HLS INTERFACE m_axi depth=160 port=output offset=slave
```
* Set the return type by double clicking on _aes_ and selecting _INTERFACE_. This time, set the return type to <i>s_axilite</i>. The tools will automatically bundle this with our other AXILite interfaces. 

### Synthesis results
Run synthesis, and scroll down to the interface summary in the report. Here you can easily see the difference between AXILite and the full AXI4 protocol, which has significantly more control signals.

From the explorer window, you can open the software drivers file, located at `solution1 > impl > misc > drivrs > aes_v1_0 > src > xaes_hw.h`. Here you can see register map for the IP, and how the interface ports have been mapped into the address space. The addresses of the input and output arrays (where the arrays are allocated in the PS DRAM) will be written to offsets 0x24 and 0x2c. The data can be read from and written to these addresses by the IP.

## Block Design in Vivado

Create a new Vivado project, selecting _RTL Project_ without specifying sources, and targeting the PYNQ-Z2. 

* Create a new block design and add the _ZYNQ7 Processing System_
* Run block automation
* In Settings, go to IP > Repository, and add the IP you exported. 
* Add the Aes IP to the Block design

* Double click on the Processing System to re-cutomize the IP
* Under _PS-PL Configuration_, expand _HP Slave AXI Interface_ and tick _S AXI HP0 Interface_. This is the port we will be connecting the <i>m_axi</i> interface from our AES IP.
* Run Connection Automation

This will instatiate 3 new blocks in the design: a _Processor System Reset_ which manages the clock and reset for the various blocks, an _AXI Interconnect_ to join the AXILite bundle (containing the key, size and return ports) on the AES IP to the master interface of the PS, and an _AXI SmartConnect_ to join the master interface of the AES IP to the High Performace slave interface of the PS.

* Validate the design to ensure everything is connected correctly
* Create the HDL wrapper for the design (right click on the design in the Sources panel)

### Instatiate the Overlay and Enable MMIO

Instantiate the overlay using the bitstream file. Make sure the .hwh or .tcl file has the same name as the .bit file.

````python
from pynq import Overlay
aes_overlay = Overlay("./bitstream/aes_master.bit")
````

As before, the key is written using MMIO, as the corresponding port uses an AXILite slave interface. 

````python
from pynq import MMIO
IP_BASE_ADDRESS = aes_overlay.ip_dict['aes_0']['phys_addr']
ADDRESS_RANGE = 0x10000
mmio = MMIO(IP_BASE_ADDRESS, ADDRESS_RANGE)
````

````python
from EncryptionAes import key_to_bytes
key  = "0x2b7e151628aed2a6abf7158809cf4f3c"
key_bytes = key_to_bytes(key)
````

### Input an image

Like with the stream interfaces, having the master interface on the PL allows multiple blocks of data to be transferred. The AXI master interface allows the IP to initiate memory transfers when required, without the need for a DMA. The IP can read the input data, and write the results back to memory according to its own schedule. 

First, we need to load an image. Note the last two lines here only serve to display the image, only the path is necessary for the next steps.

````python
import numpy as np
from PIL import Image
from IPython.display import display

image_path = "./images/tiger.bmp"
original_image = Image.open(image_path)
display(original_image)
````

We can now extract the image as a `bytearray` by opening the file in binary read mode. The bytearray is a convenient type to work with and copy to the contiguous memory buffers that were previously allocated. 

````python
with open(image_path, "rb") as image:
    f = image.read()
    b = bytearray(f)
````

Just like with the streams, we can use the [Xlnk Class](https://pynq.readthedocs.io/en/v2.4/pynq_libraries/xlnk.html) to create contiguous arrays PS DRAM. In this case however, a DMA will not manage the transfer of data, as this will be controlled directly by the IP. The image file we are reading here is a bitmap image, which has a header of 54 bytes. If we encrypt these bytes, we will not be able to display the result, as it will not be recognisable as an image file. Because of this, we are only encrypting part of `b`. 

````python
from pynq import Xlnk
xlnk=Xlnk()

header_size = 54
in_buffer = xlnk.cma_array(shape=(len(b)-header_size), dtype=np.uint8, cacheable=1)
out_buffer = xlnk.cma_array(shape=(len(b)-header_size), dtype=np.uint8, cacheable=1)
in_buffer[:(len(b)-header_size)] = b[header_size:] 
````

### Write data with a Master interface

The key and the size are written using MMIO as before, over the AXI slave interface. To write the input and output buffers, we actually pass the physical address of the arrays to the IP, over the same AXI slave interface using MMIO. This is possible thanks to the option we set in HLS to use _offset = slave_ which created additional registers in the AXI slave address space to store the address o the arrays. 

We can check the addresses of these registers either with the query `aes_overlay.ip_dict`, or by checking the software drivers file in the HLS project. 

````python
def write_data():                     
    mmio.write(0x34, len(in_buffer)*4)
    mmio.write(0x10, key_bytes)
    mmio.write(0x24, in_buffer.physical_address)  
    mmio.write(0x2c, out_buffer.physical_address)
````

Because we chose the return type <i>ap_ctrl_hs</i> in this case, we need to manually start the IP by writing the value 1 to the control register. The other signals will be handled automatically, and this register allows us to read the status of the IP at any time using `mmio.read(0x00)`.

````python
write_data()
mmio.write(0x00, 0x1)   
````

### Display the encrypted image

As mentioned when we were inputting the image data, bitmap images have a 54 byte header, which allows them to be recognised as an image and displayed correctly. For the purposes of this example, we did not encrypt this part of the original data, so we can simply concatenate the header and the encrypted output. We then open a new file in binary write mode, write the data into the file, and display as an image. 

````python
encrypted_bmp = bytes(b[:header_size])+ bytes(out_buffer)

with open("./images/encrypted_image.bmp", 'wb') as imagefile:
    imagefile.write(bytes(encrypted_bmp))

encrypted_image = Image.open("./images/encrypted_image.bmp")
display(encrypted_image)
````
